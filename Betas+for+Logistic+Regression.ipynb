```python
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2,int(num**0.5)+1):
        if num%i == 0:
            return False
    return True
n = int(input('Please enter a positive number: '))
if is_prime(n):
    print('number entered is prime')
else:
    print('number entered is not prime')
        
```

```python
print(f"Test accuracy score: {accuracy_score(y_test,prediction_oneVsrest)*100}%\n\n")
print(f"Classification Report: \n\n{classification_report(y_test,prediction_oneVsrest)}")
```
